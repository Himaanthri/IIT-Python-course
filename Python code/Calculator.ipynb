# Menu
1 - Addition 
2 - Subtraction 
3 - Multiplication 
4 - Division 
5 - Exit

```python
menu = 1
while menu!=5:
    menu = int(input("Choose the option"))
    if menu == 1:
        a = int(input("Enter a number"))
        b = int(input("Enter a number"))
        print("Addition:",a+b)
    elif menu == 2:
        a = int(input("Enter a number"))
        b = int(input("Enter a number"))
        print("Subtraction:",a-b)
    elif menu == 3:
        a = int(input("Enter a number"))
        b = int(input("Enter a number"))
        print("Multiplication:",a*b)
    elif menu == 4:
        a = int(input("Enter a number"))
        b = int(input("Enter a number"))
        if b==0:
            print("Invalid")
            pass
        else:
            print("Division:",a/b)
    elif menu != 5:
        print("Invalid")
        pass
print("Good Bye")
```
